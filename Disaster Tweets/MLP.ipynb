```python
## Imports Absolus
import argparse
import os
import string
from numpy import array, argmax, random, take
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
import emoji
import itertools
import nltk
import collections
from collections import Counter
from tokenizers.processors import BertProcessing
from tokenizers import (ByteLevelBPETokenizer,
                            CharBPETokenizer,
                            SentencePieceBPETokenizer,
                            BertWordPieceTokenizer)
# Keras
from keras.models import Sequential
from keras.layers import Dense, Activation,LSTM, Embedding, Bidirectional, RepeatVector, TimeDistributed, Dropout
from keras.preprocessing.text import Tokenizer
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau,EarlyStopping
from keras.preprocessing.sequence import pad_sequences
from keras.models import load_model
from keras import optimizers
```

```python
train = pd.read_csv('concat_train.csv')
```

```python
train_values = train
```

```python
train_values = train_values.drop("target",axis=1)
```

```python
def number_keyword(text):
    if type(text)==float:
        return 0
    else:
        text_list = str(text).split("%")
        return len(text_list)

train_values['keyword']=train_values['keyword'].apply(lambda x : number_keyword(x))
```

```python
train_values
```

```python
train_values = train_values.drop("id",axis=1)
#train_values = train_values.drop("Score_Sentence",axis=1)
#train_values = train_values.drop("sentiment_rate",axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_sub_train, x_sub_test, y_sub_train, y_sub_test = train_test_split(train_values,train['target'], test_size=0.3, random_state=42)
```

```python
train_values
```

```python
checkpoint = ModelCheckpoint("MLP_Deep_Dumb_MLP_2", monitor='val_acc', verbose=1, save_best_only=True, mode='min')
```

```python
input_dim = 7
nb_classes = 2
```

```python
# Here's a Deep Dumb MLP (DDMLP)
model = Sequential()
model.add(Dense(128, input_dim=input_dim))
model.add(Activation('relu'))
model.add(Dropout(0.15))
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.15))
model.add(Dense(32))
model.add(Activation('relu'))
model.add(Dropout(0.15))
model.add(Dense(nb_classes))
model.add(Activation('softmax'))

# we'll use categorical xent for the loss, and RMSprop as the optimizer
model.compile(loss='sparse_categorical_crossentropy', optimizer='rmsprop')
```

```python
history=model.fit(x_sub_train, y_sub_train,
          batch_size=24,
          epochs=100,
          validation_data=[x_sub_test, y_sub_test],
          callbacks=[checkpoint],
          verbose=1) 
```

```python
preds = model.predict_classes(x_sub_test, verbose=0)
```

```python
x_sub_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

predicted_classes = preds
y_test = y_sub_test

test_scores = accuracy_score(y_test,predicted_classes), precision_score(y_test, predicted_classes), recall_score(y_test, predicted_classes), f1_score(y_test, predicted_classes)

print('\n \n Scores')
scores = pd.DataFrame(data=[test_scores])
scores.columns = ['accuracy', 'precision', 'recall', 'f1']
scores = scores.T
scores.columns = ['scores']
display(scores)
```
